```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("emails.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df.drop(columns=['Prediction', 'Email No.'])
y = df['Prediction']
```

```python
X
```

```python
X = df.drop(columns=['Prediction', 'Email No.'])
y = df['Prediction']

spam = (y == 1).sum()  
notspam = (y == 0).sum()  

print("Spam emails:", spam)
print("Normal emails:", notspam)
        
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LogisticRegression(max_iter = 300, C=1.5)
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("the logistic regression's predictions on test set were, ",accuracy*100,"% accurate.")
```

```python
import pandas as pd
import nltk
import string
from collections import Counter

nltk.download('punkt_tab')
```

```python
email_text = """
Subject: Exclusive Offer! Unlock Your Reward Now!

Dear Valued Customer,

Congratulations! You are one step away from unlocking an exclusive reward just for YOU. To receive your prize, simply click the link below and follow the easy instructions.

This is a limited-time offer for our valued members, so act now before it's too late!

By clicking, you'll have access to exclusive deals, top-tier infrastructure, and more. Our system is allowing lucky users like you to claim amazing benefits instantly!

Here’s what you need to do:

Click the link and sign in.
Provide your details to verify your identity.
Claim your reward.
Don’t wait! The military-grade encryption ensures your information is secure. We look forward to providing you with the best offers.

Sincerely,
The Rewards Team
valued@exclusiveoffers.com
"""
tokens = nltk.word_tokenize(email_text.lower())
tokens = [word for word in tokens if word.isalnum()]  # Remove punctuation and non-alphanumeric

# Count occurrences of each word in the tokenized email
word_counts = Counter(tokens)

email_row = {}

for col in df.columns:
    if col not in ['Email No.', 'Prediction']:
        email_row[col] = word_counts.get(col, 0)

email_row['Email No.'] = len(df) + 1  # Auto-generate email number
email_row['Prediction'] = None  # Placeholder, can be spam/ham or prediction result
```

```python
test_email = pd.DataFrame(columns = df.columns)
```

```python
test_email= test_email._append(email_row,ignore_index=True)
```

```python
test_email = test_email.drop(columns=['Prediction', 'Email No.'])
```

```python
scaler.fit_transform(test_email)
```

```python
prediction = model.predict(test_email)
```

Let us try different models:


```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()

X_train[X_train < 0] = 0 # handling negative data, as MultinomialNB deals only with +ve data
nb.fit(X_train, Y_train)

```

```python
y_pred = nb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("the model's predictions on test set were, ",accuracy*100,"% accurate.")
```

lmaoo, logistic regression was better 

LET'S TRY XGBoost!

```python
from xgboost import XGBClassifier
xbg_model = XGBClassifier(n_estimators = 300, max_depth = 4)

xbg_model.fit(
    X_train, Y_train,
    eval_set = [(x_test, y_test)],
    verbose = 1,

)

```

```python
xbg_model.score(x_test, y_test)
```

XGBoost did 1% better than Logistic Regression on the test set!
